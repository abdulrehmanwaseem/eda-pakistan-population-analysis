### **Overview**
This dataset contains demographic information from the **Pakistan Population Census 2017**.  
It provides detailed population data at various administrative levels within Pakistan, including **provinces, divisions, districts, and sub-divisions**.  

The dataset also includes insights into **urban/rural populations, gender distribution, transgender individuals, sex ratios, historical comparisons (1998), and annual growth rates**.

---

### **Features**

- **Province**: Administrative provinces/regions of Pakistan where the census data was collected.  
- **Division**: The divisions within each province (2nd-level administrative units).  
- **District**: Districts within each division (larger administrative units).  
- **Sub-Division**: Sub-divisions/tehsils within each district (localized level).  
- **Area (sq. km)**: Land area of each sub-division.  
- **Urban Population (2017)**: Population living in urban areas.  
- **Rural Population (2017)**: Population living in rural areas.  
- **Male Population (2017)**: Male population count.  
- **Female Population (2017)**: Female population count.  
- **Transgender Population (2017)**: Transgender population count.  
- **Sex Ratio (2017)**: Number of females per 1000 males.  
- **Population (1998)**: Population count from the 1998 census.  
- **Annual Growth Rate (%)**: Annual population growth rate between 1998â€“2017.

```python
pop_rural_2017 = df["ALL SEXES (RURAL)"].sum()
pop_rural_1998 = df["POPULATION 1998 (RURAL)"].sum()
```

```python
print("The total rural population in 1998 was:", pop_rural_1998)
print("The total rural population in 2017 was:", pop_rural_2017)
```

```python
# How much rural population has changed from 1998 to 2017?
rural_change = pop_rural_2017 - pop_rural_1998
if rural_change > 0:
    print("The rural population increased by:", rural_change)
    print(f"The rural population increased in percentage: {round(rural_change / pop_rural_1998 * 100, 2)}%")
else:
    print("The rural population decreased by:", abs(rural_change))
    print(f"The rural population decreased in percentage: {round(abs(rural_change) / pop_rural_1998 * 100, 2)}%")
```

```python
pop_urban_2017 = df["ALL SEXES (URBAN)"].sum()
pop_urban_1998 = df["POPULATION 1998 (URBAN)"].sum()

# How much urban population has changed from 1998 to 2017?
urban_change = pop_urban_2017 - pop_urban_1998
if urban_change > 0:
    print("The urban population increased by:", urban_change)
    print(f"The urban population increased in percentage: {round(urban_change / pop_urban_1998 * 100, 2)}%")
else:
    print("The urban population decreased by:", abs(urban_change))
    print(f"The urban population decreased in percentage: {round(abs(urban_change) / pop_urban_1998 * 100, 2)}%")
```

```python
print("The difference btw urban and rural population is:", abs(urban_change - rural_change))
```

```python
sns.barplot(x=["Rural Population Change", "Urban Population Change"], y=[rural_change, urban_change])
```
